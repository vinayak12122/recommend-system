```python
import pandas as pd

df = pd.read_csv("Dataset/smartphone.csv")
df.head()
```

```python
df["processor_brand"].unique()
```

```python
df.isnull().sum()
df
```

```python
df['rating'].describe()

df['rating'].fillna(df['rating'].median(),inplace=True)
df.isnull().sum()
```

```python
df
```

```python
mapping = {
    'Octa': 8,
    'Hexa': 6,
    'Quad': 4,
    'Dual': 2
}

# This extracts the key words and maps them to numbers in one go
df['num_cores'] = df['num_cores'].str.extract('(Octa|Hexa|Quad|Dual)', expand=False).map(mapping)

df.drop(columns=['processor_name'],inplace=True)
df['processor_brand'].fillna(df['processor_brand'].mode()[0],inplace=True)
df['num_cores'].fillna(df['num_cores'].median(),inplace=True)
df['processor_speed'].fillna(df['processor_speed'].median(),inplace=True)
df['battery_capacity'].fillna(df['battery_capacity'].mode()[0],inplace=True)
df['ram_capacity'].fillna(df['ram_capacity'].median(),inplace=True)
df['internal_memory'].fillna(df['internal_memory'].mode()[0],inplace=True)
df['resolution'].fillna(df['resolution'].mode()[0],inplace=True)
df['os'].fillna(df['os'].mode()[0], inplace=True)
```

```python
df
```

```python
df['num_cores'].unique()
df.isnull().sum()
```

```python
df.to_csv('Dataset/new_phone.csv',index=False)
```
